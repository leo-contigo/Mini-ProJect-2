```python
# 경고 메세지 숨기기 
import warnings
warnings.filterwarnings(action='ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


import plotly.express as px
# https://plotly.com/python/plotly-fundamentals/

import plotly.graph_objects as go

# 그래프 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
# plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['font.size'] = 16
plt.rcParams['figure.figsize'] = 20, 10

# 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
plt.rcParams['axes.unicode_minus'] = False

# 글꼴 선명화 
%config InlineBackend.figure_format = 'retina'


df = pd.read_csv('./제주특별자치도_주제8_유동인구 비율 대비 지역별 인기있는 업종 추천 데이터_매쉬업결과_20201215.csv', encoding='cp949')
df.drop(['데이터기준일자', '총 유동인구'], axis=1, inplace=True)


# 업종명 면세점 찾기
df[df['업종명'] == "면세점"]
# 면세점에 해당하는 인덱스
df[df['업종명'] == "면세점"].index
del_1 = df[df['업종명'] == "면세점"].index
# 행 삭제하기
df.drop(del_1, inplace=True)

# 업종명 정기 항공 운송업 찾기
df[df['업종명'] == "정기 항공 운송업"]
# 정기 항공 운송업 해당하는 인덱스
df[df['업종명'] == "정기 항공 운송업"].index
del_1 = df[df['업종명'] == "정기 항공 운송업"].index
# 행 삭제하기
df.drop(del_1, inplace=True)

# 업종명 정기 항공 운송업 찾기
df[df['업종명'] == "기타 갬블링 및 베팅업"]
# 정기 항공 운송업 해당하는 인덱스
df[df['업종명'] == "기타 갬블링 및 베팅업"].index
del_1 = df[df['업종명'] == "기타 갬블링 및 베팅업"].index
# 행 삭제하기
df.drop(del_1, inplace=True)

a100 = df.groupby(['시도명','업종명']).mean() // 10000
a100.reset_index(inplace=True)
a100 = a100[['시도명','업종명','이용금액']]

fig = px.sunburst(a100, width=750, height=750,
            path=['시도명','업종명', '이용금액'],
            maxdepth=2,
            values ='이용금액',
            color='이용금액',
            color_continuous_scale='rdbu'
)

fig.update_traces(branchvalues='total', selector=dict(type='sunburst'), 
                  hovertemplate= '시도명=%{parent}<br><br>업종명=%{label}<br><br>이용금액=%{value}(만원)<br><extra></extra>'
)

# 그래프 제목 설명
fig.update_layout(
        title=dict(
        # <br> 태크와 <sup>태그 사용해서 서브 타이틀을 작성할 수 있음
        text='<b>제주 업종별 평균 매출액</b><br><sup>( 기간 : 2018년 ~ 2020년 4월까지 / 금액단위 : (만원)</sup>',
        x=0.5,
        font=dict(
            family="Malgun Gothic",
            size=22,
            color="#000000"
        )))

fig.show()
```

```python
# html 저장하기
# f = open("파일경로.html","w")

a = fig.to_html()
f = open("a.html","w")
f.write(a)
f.close()
```
